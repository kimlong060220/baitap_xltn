```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

## Loading audio files with Librosa

```python
audio_file = "D:\\xltn\\dataset\\a\\file1.wav"
```

```python
ipd.Audio(audio_file)
```

```python
# load audio files with librosa
signal, sr = librosa.load(audio_file)
```

## Extracting MFCCs

```python
# more parameters from librosa.feature.melspectrogram
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=13, sr=sr)
mfccs.shape
```

```python
# Computing delta
delta_mfccs = librosa.feature.delta(mfccs)
delta_mfccs.shape
```

```python
## Computing deltadelta
deltadelta_mfccs = librosa.feature.delta(mfccs, order=2)
deltadelta_mfccs.shape
```

```python
# Concat mfcc , delta and deltadelta
mfcc = np.concatenate((mfccs, delta_mfccs, deltadelta_mfccs))
```
